```python
from nltk.corpus import stopwords
from collections import Counter
import pandas as pd
import string
import nltk
import re

nltk.download("stopwords")

stop = set(stopwords.words("english"))

# functions for pre Processing
def remove_comma(text):
    return text.replace("," ," ")  

# remove url or www .com etc...
def remove_url(text):
    url = re.compile(r"https?://\S+www\.\S+")
    return url.sup(r"" ,text)    

# remove /?}{ \ etc....
def remove_punc (text):
    translator = str.maketrans("","" ,string.punctuation)
    return text.translate(translator)

# remove an,a,the,on,that,this ,etc....
def remove_stopwords(text):
    filered_word = [word.lower() for word in text.split() if word.lower() not in stop]
    return " ".join(filered_word)



def counter_words(text_col):
    count = Counter()
    
    for i in range(len(text_col)):
        for word in text_col[i].split():
            count[word] += 1
    return count

def counter_list_iter (list_):
    counter =  Counter()
    for i in range(0,len(list_)):
        counter.update(counter_words(list_[i].split()))
    return counter

```

```python
counter =  Counter()

for i in range(0,len(df)):
    text = df["description"][i]
    
    text = remove_comma(text)
    # text = remove_url(text)
    text = remove_punc(text)
    text = remove_stopwords(text)
    counter.update(counter_words(text.split()))


print(counter.most_common)
print(len(counter))
```
